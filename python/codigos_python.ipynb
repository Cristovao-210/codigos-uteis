# Colocar nome no padrão SIAPE

```python
def nome_padrao_siape(txt):

    for ch in txt:
      match ch:
        case 'à':
          txt = txt.replace(ch, 'a')
        case 'á':
          txt = txt.replace(ch, 'a')
        case 'ã':
          txt = txt.replace(ch, 'a')
        case 'â':
          txt = txt.replace(ch, 'a')
        case 'é':
          txt = txt.replace(ch, 'e')
        case 'ê':
          txt = txt.replace(ch, 'e')
        case 'í':
          txt = txt.replace(ch, 'i')
        case 'ó':
          txt = txt.replace(ch, 'o')
        case 'ô':
          txt = txt.replace(ch, 'o')
        case 'õ':
          txt = txt.replace(ch, 'o')
        case 'ú':
          txt = txt.replace(ch, 'u')
        case 'ü':
          txt = txt.replace(ch, 'u')
        case 'ç':
          txt = txt.replace(ch, 'c')

        # no caso das letras serem maiúsculas
        case 'À':
          txt = txt.replace(ch, 'A')
        case 'Á':
          txt = txt.replace(ch, 'A')
        case 'Ã':
          txt = txt.replace(ch, 'A')
        case 'Â':
          txt = txt.replace(ch, 'A')
        case 'É':
          txt = txt.replace(ch, 'E')
        case 'Ê':
          txt = txt.replace(ch, 'E')
        case 'Í':
          txt = txt.replace(ch, 'I')
        case 'Ó':
          txt = txt.replace(ch, 'O')
        case 'Ô':
          txt = txt.replace(ch, 'O')
        case 'Õ':
          txt = txt.replace(ch, 'O')
        case 'Ú':
          txt = txt.replace(ch, 'U')
        case 'Ü':
          txt = txt.replace(ch, 'U')
        case 'Ç':
          txt = txt.replace(ch, 'C')

    return txt.upper()
```

# Algoritimos de Ordenação



```python
# Selection sort

lista2 =  [60,73,300,15,-1,86,11,9,4,1,0]
print(lista2)

for pos in range(len(lista2)):
  menor = pos
  for pos1 in range(pos + 1, len(lista2)):
    if lista2[pos1] < lista2[menor]:
      menor = pos1
  aux2 = lista2[pos]
  lista2[pos] = lista2[menor]
  lista2[menor] = aux2

print(lista2)

# Bubble sort

lista = [14, 5, 11, 9, 1, 450]
print(lista)

for i in range(len(lista)):
  for num in range(1, len(lista)):
    if lista[num-1] > lista[num]:
      aux = lista[num-1]
      lista[num-1] = lista[num]
      lista[num] = aux

print(lista)
```

# Switch case Python (antes da função 'match')

```python
# Switch Cases
# O Python NÃO tem uma implementação para os switch cases, mas uma maneira de implementá-lo é com o dicionário, uma estrutura de dados que armazena o par de valores chave (Módulo 3).

# As condições switch são armazenadas como chaves no dicionário, e as ações armazenadas como valor.
# Se houver uma série de ações para cada caso, então considere escrever uma função para cada caso e use as chamadas de função como o valor.
# A condição padrão é listada manualmente como um valor chave no get().
def janeiro():
  return "janeiro"

def fevereiro():
  return "fevereiro"

def switcher(number):
  # Use dicionário (do Módulo 3) para armazenar switch cases
  # Se não for encontrado, o get() será o valor padrão
  return {
    '1': janeiro(),
    '2':fevereiro(),
    '3':"março",
    '4':"abril",
    '5':"maio",
    '6':"junho",
    '7':"julho",
    '8':"agosto",
    '9':"setembro",
    '10':"outubro",
    '11':"novembro",
    '12':"dezembro"
  }.get(number,"Número de mês inválido!")


# input() lê uma entrado do usuário de stdin
number = input("Informe o número do mês desejado: ")
switcher(number)
```

```python
# imports necessários
import datetime

# função para deixar data recebida no formato necessário
def data_recebida(dt): # recebe uma String
  dia = dt[8:]
  mes = dt[5:7]
  ano = dt[0:4]
  return datetime.date(int(ano), int(mes), int(dia)) # retorna tipo datetime

# função para deixar data recebida no formato usado no Brasil
def format_dt_br(dat): # recebe um objeto dateTime
  dia = (f'0{str(dat.day)}' if int(dat.day) < 10 else str(dat.day))
  mes = (f'0{str(dat.month)}' if int(dat.month) < 10 else str(dat.month))
  ano = str(dat.year)
  return f'{dia}/{mes}/{ano}' # retorna uma String

# função para calcular períodos entre uma data e outra
def calcula_perido(d_ini, d_fim):
  dif_periodo = data_recebida(d_fim) - data_recebida(d_ini)
  dias_periodo = (dif_periodo.days + 1)
  return dias_periodo

# função para gerar sequência de anos
#(ou qualquer outra sequẽncia numérica progressiva)
def define_anos(a_ini, a_fim):
  lista_anos = [a_ini]
  for ano in range((a_fim - a_ini)):
    lista_anos.append((a_ini + (ano + 1)))
  return lista_anos

# função para retirar anos que não devem constar no(s) período(s)
def retira_ano(a_retirar, a_periodo):
  anos_informados = a_retirar.split(",")
  for year in anos_informados:
    if int(year) in a_periodo:
      a_periodo.remove(int(year))
  return a_periodo

# função para calcular anos, meses e dias
"""
FALTA DEFINIR CRITÉRIOS DE ARREDONDAMENTO E SE O
MÊS VAI SER DE 30 DIAS OU DE 30.4
"""
def calcula_a_m_d(tempo_dias):
  ANO = 365
  MES = 30
  anos = tempo_dias // ANO
  resto_anos = tempo_dias % ANO
  meses = resto_anos // MES
  dias = resto_anos % MES
  return f'{int(anos)} ano(s), {int(meses)} mês(es), {int(dias)} dia(s)'

periodo = int(input("Informe a quantidade de dias a ser calculada: "))
resultado = calcula_a_m_d(periodo)
print(resultado)

# possível correção para o "bug dos 364 dias"
tempo_dias = int(input("Informe a quantidade de dias a ser calculada: "))
ANO = 365
MES = 30.4
anos = tempo_dias // ANO
resto_anos = tempo_dias % ANO
meses = resto_anos // MES
dias = resto_anos % MES
result = f'{int(anos)} ano(s), {int(meses)} mês(es), {int(dias)} dia(s)'
print(result)

hoje = datetime.date.today()
julho2012 = datetime.date(2012,7,22)
intervalo_tempo = hoje - julho2012
intervalo_tempo.days
```

# Cálculo do Auxílio Transponte Servidor



```python
# calculando o valor de um percurso (ida ou volta)
def calc_percurso(qtd_percursos, tipo_percurso):
  total_trecho = 0
  for perc in range(qtd_percursos):
    valor = float(input(f'Informe o valor do percurso {perc + 1} de {tipo_percurso}: '))
    total_trecho += valor
  return total_trecho

# calculando o valor mensal com base nos percursos informados

# valores constantes
D_6_P_C = 0.06
MES = 30
DIAS_TRABALHO = 22

# ida
num_percursos = int(input("Informe a quantidade de percursos utilizados para ida ao trabalho: "))
valor_ida = calc_percurso(num_percursos, "ida")

# volta
num_percursos = int(input("Informe a quantidade de percursos utilizados para volta do trabalho: "))
valor_volta = calc_percurso(num_percursos, "volta")

# Total percursos diário
valor_diario = (valor_ida + valor_volta) * DIAS_TRABALHO

# recebendo valor vencimento básico
venc_basico = float(input("Informe o valor do seu vencimento básico: "))

# calculando valor a ser descontado (6%)
valor_desconto = (venc_basico / MES) * DIAS_TRABALHO * D_6_P_C

# calculando valor a ser pago pelo servidor
valor_pago = valor_diario - valor_desconto

print(f'-----------------------------------------------\nValor diário: R$ {valor_diario}\nValor descontado (6%): R$ {valor_desconto}\nValor do Auxílio Pago: R$ {valor_pago}')
```
